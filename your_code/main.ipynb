#### What's the mean salary by department?
Hint : There are 2 ways to do it - one way is with groupby and the other way is with pivot tables. Do it both ways. 


```python
employee.groupby(['Department']).agg({'Salary' : np.mean})
```

#### We want now to group the data by more than one column. Use Pandas pivot table function to select the mean salary by department and title.

```python
employee.pivot_table(index=['Department'], columns=['Title'], values= ['Salary']) # The mean is calculated by default
```

```python
employee.pivot_table(index=['Department', 'Title'], values= ['Salary'])
```

#### We want to display the number of years of work experience and the salary for each Title. Add Years as a column in your pivot table.
*Hint: Use Pandas doc(see References in README.md) to explore the pivot_table() arguments*

```python

employee.pivot_table(index= ['Department','Title'], values = ['Salary','Years'],aggfunc ={'Salary':'mean','Years':'mean'})
```

#### We want to know the information about the mean salary and number of employees in each department and for each job title.

```python
employee.pivot_table(index=['Department', 'Title'], values = ['Salary','Name'],aggfunc ={'Salary':'mean','Name':'nunique'} )
```

#### We want to know for each department and for each job title the Salary's median, minimum, maximum and standard deviation. Comment your results.

```python
describe_employee = employee.pivot_table (index= ['Department','Title'], values = ['Salary'],aggfunc =['median', 'max', 'min', 'std'])
describe_employee
# In the results we can see that some job titles present NaN values in standard deviation column, this is due to the fact 
# that it is being calculated for only one employee in the job title category, and therefore there is no standard deviation
# when there is just one single value. On the other hand 'Sales associate' title has an std of 0 because there are two 
# employees with the same salary and therefore there is no standard deviation between identical values.
```

#### Based on your comments, fill the missing values with an appropriate value.

```python
#describe_employee.isna().sum()
correct_std = describe_employee['max'] - describe_employee['median']

describe_employee['std'] = describe_employee['std'].fillna(correct_std)
describe_employee
```

#### The stake holders want to know for each department the number of employees and how much money is spend on salaries. Could you provide that information? 

```python
employee.pivot_table(index=['Department'], aggfunc={'Name':'nunique','Salary':'sum'})
```

#### For each Department and Title, we want to know the the total years of work experience and the mean salary.

```python
employee.pivot_table(index=['Department','Title'], aggfunc={'Years':'sum','Salary':'mean'})
```
